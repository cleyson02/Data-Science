**Sobre a base de dados**

Os dados de rotatividade de clientes de telecomunicações contêm informações sobre uma empresa de telecomunicações fictícia que forneceu serviços de telefone residencial e Internet para 7.043 clientes na Califórnia.

**Dicionário de dados**


* CustomerId: Customer Id
* Gender: Gender
* SeniorCitizen: Whether the customer is elderly (1.0)
* Partner: Whether the customer has a partner (Yes, No)
* Dependents: Whether the customer has dependents (Yes, No) (child, mother, father...)
* tenure: The number of months the customer stays in the company
* PhoneService: Whether the customer has phone service (Yes, No)
* MultipleLines: Whether the customer has more than one line (Yes, No, No phone service)
* InternetService: Customer's internet service provider (DSL, Fiber optic, No)
* OnlineSecurity: Whether the customer has online security (Yes, No, no Internet service)
* OnlineBackup: Whether the customer has an online backup (Yes, No, no Internet service)
* DeviceProtection: Whether the customer has device protection (Yes, No, no Internet service)
* TechSupport: Whether the customer receives technical support (Yes, No, no Internet service)
* StreamingTV: Whether the customer has broadcast TV (Yes, No, no Internet service)
* StreamingMovies: Whether the customer is streaming movies (Yes, No, no Internet service)
* Contract: Customer's contract period (Month to month, One year, Two years)
* PaperlessBilling: Whether the customer has a paperless invoice (Yes, No)
* PaymentMethod: Customer's payment method (Electronic check, Postal check, Bank transfer, Credit card)
* MonthlyCharges: Amount collected from the customer monthly
* TotalCharges: Total amount collected from the customer
* Churn: Yes = the customer left the company this Month. No = the customer remained with the company.

```python
import pandas as pd

path = 'https://raw.githubusercontent.com/profEraylson/datasets/main/churn_telecom.csv'

df = pd.read_csv(path)

df
```

```python
# Qual a porcentagem de clientes que cancelaram o serviço (churn)?

clientes_churn = df['Churn'].value_counts()['Yes']

clientes_total = len(df)

resultado = (clientes_churn / clientes_total) * 100

print("Porcentagem de clientes que cancelaram o serviço (churn):", resultado, "%")
```

```python
# Qual a distribuição dos clientes por gênero, idade (SeniorCitizen), estado civil (Partner), dependentes (Dependents)?

genero = df['gender'].value_counts()
print("Distribuição por Gênero:")
print("Número de clientes do sexo feminino:", genero['Female'])
print("Número de clientes do sexo masculino:", genero['Male'])
print()

idade = df['SeniorCitizen'].value_counts()
print("Distribuição por Idade:")
print("Número de clientes que não são idosos:", idade[0])
print("Número de clientes que são idosos:", idade[1])
print()

estado_civil = df['Partner'].value_counts()
print("Distribuição por Estado Civil:")
print("Número de clientes sem parceiro:", estado_civil['No'])
print("Número de clientes com parceiro:", estado_civil['Yes'])
print()

dependentes = df['Dependents'].value_counts()
print("Distribuição por Dependentes:")
print("Número de clientes sem dependentes:", dependentes['No'])
print("Número de clientes com dependentes:", dependentes['Yes'])
```

```python
# Qual a média, mediana e desvio padrão da permanência na empresa (tenure) e dos valores mensais (MonthlyCharges) e totais (TotalCharges)?

# Convertendo a coluna "TotalCharges" para float
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Preenchendo os valores ausentes com a média da coluna "TotalCharges"
media_TotalCharges = df['TotalCharges'].mean()
df['TotalCharges'] = df['TotalCharges'].fillna(media_TotalCharges)

# Verificando o tipo da coluna após a conversão
print("Tipo da coluna 'TotalCharges' após a conversão:", df['TotalCharges'].dtype)
print()

resultado = df[['tenure', 'MonthlyCharges', 'TotalCharges']].agg(['mean', 'median', 'std'])

print(resultado)
```

```python
# Clientes do gênero masculino ou feminino têm maior propensão ao churn?

clientes_genero = df['gender'].value_counts()

clientes_churn = df.groupby('gender')['Churn'].value_counts()

taxa_female = clientes_churn['Female']['Yes'] / clientes_genero['Female']
taxa_male = clientes_churn['Male']['Yes'] / clientes_genero['Male']

print("Taxa de churn para clientes do sexo feminino:", taxa_female)
print("Taxa de churn para clientes do sexo masculino:", taxa_male)

if taxa_female > taxa_male:
    print("Clientes do sexo feminino têm maior propensão ao churn.")
elif taxa_female < taxa_male:
    print("Clientes do sexo masculino têm maior propensão ao churn.")
else:
    print("Não há diferença significativa na propensão ao churn entre os gêneros.")
```

```python
# Clientes idosos (SeniorCitizen) tendem a cancelar o serviço com mais frequência?

total_senior = df['SeniorCitizen'].value_counts()

churn_senior = df.groupby('SeniorCitizen')['Churn'].value_counts()

taxa_senior = churn_senior[1]['Yes'] / total_senior[1]
taxa_no_senior = churn_senior[0]['Yes'] / total_senior[0]

print("Taxa de churn para clientes idosos:", taxa_senior)
print("Taxa de churn para clientes não idosos:", taxa_no_senior)

if taxa_senior > taxa_no_senior:
    print("Clientes idosos tendem a cancelar o serviço com mais frequência.")
elif taxa_senior < taxa_no_senior:
    print("Clientes não idosos tendem a cancelar o serviço com mais frequência.")
else:
    print("Não há diferença significativa na propensão ao churn entre clientes idosos e não idosos.")
```

```python
# Clientes casados (Partner) ou com dependentes (Dependents) apresentam taxas de churn diferentes?

clientes_partner = df['Partner'].value_counts()

churn_partner = df.groupby('Partner')['Churn'].value_counts()

taxa_partner = churn_partner['Yes']['Yes'] / clientes_partner['Yes']

print("Taxa de churn para clientes com parceiro:", taxa_partner)

clientes_dependents = df['Dependents'].value_counts()

churn_dependents = df.groupby('Dependents')['Churn'].value_counts()

taxa_dependents = churn_dependents['Yes']['Yes'] / clientes_dependents['Yes']

print("Taxa de churn para clientes com dependentes:", taxa_dependents)

import matplotlib.pyplot as plt

categorias = ['Clientes Casados', 'Clientes com Dependentes']
taxas_churn = [taxa_partner, taxa_dependents]

plt.bar(categorias, taxas_churn)
plt.title('Taxa de Churn entre Clientes Casados e com Dependentes')
plt.ylabel('Taxa de Churn')
plt.show()
```

```python
# Existe alguma relação entre o número de linhas telefônicas (MultipleLines) e o churn?

import matplotlib.pyplot as plt
import seaborn as sns

churn_taxa_multiplelines = df.groupby('MultipleLines')['Churn'].value_counts(normalize=True).unstack()

churn_taxa_multiplelines.plot(kind='bar')
plt.title('Taxa de Churn por Categoria de Linhas Telefônicas')
plt.xlabel('MultipleLines')
plt.ylabel('Taxa de Churn')
plt.show()
```

```python
# Clientes com internet de fibra óptica (Fiber optic) cancelam menos que os demais?

churn_taxa_fiberoptic = df[df['InternetService'] == 'Fiber optic']['Churn'].value_counts(normalize=True)['Yes']

churn_taxa_outros = df[df['InternetService'].isin(['DSL', 'No'])]['Churn'].value_counts(normalize=True)['Yes']

print("Taxa de churn para clientes com internet de fibra óptica (Fiber optic):", churn_taxa_fiberoptic)
print("Taxa de churn para clientes com outros tipos de internet (DSL ou sem serviço de internet):", churn_taxa_outros)

if churn_taxa_fiberoptic < churn_taxa_outros:
    print("Clientes com internet de fibra óptica cancelam menos que os demais.")
elif churn_taxa_fiberoptic > churn_taxa_outros:
    print("Clientes com internet de fibra óptica cancelam mais que os demais.")
else:
    print("Clientes com internet de fibra óptica cancelam na mesma proporção que os demais.")
```

```python
# Serviços adicionais como segurança online (OnlineSecurity), backup online (OnlineBackup) e proteção de dispositivos (DeviceProtection) afetam a taxa de churn?

# Calculando a taxa de churn para cada categoria de serviço adicional
churn_taxa_services = df.groupby(['OnlineSecurity', 'OnlineBackup', 'DeviceProtection'])['Churn'].value_counts(normalize=True).unstack()

# Exibindo as taxas de churn para cada categoria de serviço adicional
print("Taxa de Churn por Serviços Adicionais:")
print()
print(churn_taxa_services)
```

```python
# O tipo de contrato (Month to month, One year, Two years) influencia na permanência do cliente?

import matplotlib.pyplot as plt

churn_taxa_contrato = df.groupby('Contract')['Churn'].value_counts(normalize=True).unstack()

churn_taxa_contrato.plot(kind='bar')
plt.title('Taxa de Churn por Tipo de Contrato')
plt.xlabel('Tipo de Contrato')
plt.ylabel('Taxa de Churn')
plt.show()
```

```python
# Qual a proporção de clientes que possui telefone fixo (PhoneService)?

clientes_tel = df['PhoneService'].value_counts()['Yes']

proporcao_clientes = (clientes_tel / len(df)) * 100

print("Proporção de clientes que possuem telefone fixo (PhoneService):", proporcao_clientes, "%")

import matplotlib.pyplot as plt

labels = ['Clientes com telefone', 'Clientes sem telefone']
sizes = [clientes_tel, len(df) - clientes_tel]
plt.pie(sizes, labels=labels, autopct='%.2f%%')
plt.show()
```

```python
# Quantos clientes possuem internet (InternetService) e qual a distribuição entre os tipos (DSL, Fiber optic)?

internet_customers = df['InternetService'].value_counts()

print("Número total de clientes com serviço de internet:", internet_customers.sum())
print()
print("Distribuição entre os tipos de serviço de internet:")
print(internet_customers)
```

```python
# Qual a porcentagem de clientes que utiliza serviços de streaming (StreamingTV, StreamingMovies)?

clientes_com_streaming = ((df['StreamingTV'] == 'Yes') | (df['StreamingMovies'] == 'Yes')).sum()

porcentagem_streaming = (clientes_com_streaming / len(df)) * 100

print("Porcentagem de clientes que utiliza serviços de streaming (StreamingTV, StreamingMovies):", porcentagem_streaming, "%")
```

```python
# Existe alguma relação entre o valor mensal faturado (MonthlyCharges) e o churn?

import matplotlib.pyplot as plt

monthly_charges = df.groupby('Churn')['MonthlyCharges'].describe()

print("Estatísticas resumidas dos valores mensais faturados:")
print(monthly_charges)

sns.histplot(data=df, x='MonthlyCharges', hue='Churn', kde=True)
plt.title('Distribuição dos Valores Mensais Faturados por Churn')
plt.xlabel('Monthly Charges')
plt.ylabel('Frequência')
plt.show()
```

```python
# O valor total faturado (TotalCharges) está associado a propensão do cliente cancelar o serviço?

import matplotlib.pyplot as plt

# Convertendo a coluna "TotalCharges" para float
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Preenchendo os valores ausentes com a média da coluna "TotalCharges"
media_TotalCharges = df['TotalCharges'].mean()
df['TotalCharges'] = df['TotalCharges'].fillna(media_TotalCharges)

# Verificando o tipo da coluna após a conversão
print("Tipo da coluna 'TotalCharges' após a conversão:", df['TotalCharges'].dtype)
print()

mean_TotalCharges = df.groupby('Churn')['TotalCharges'].mean()

mean_TotalCharges.plot(kind='bar')
plt.title('Comparação dos Valores Médios Totais Faturados entre Clientes que Cancelaram e não Cancelaram o Serviço')
plt.xlabel('Churn')
plt.ylabel('Média Total Charges')
plt.show()
```

```python
# Qual o método de pagamento mais utilizado pelos clientes (Electronic check, Postal check, Bank transfer, Credit card)?

metodo_pagamento_mais_utilizado = df['PaymentMethod'].value_counts().idxmax()

print("O método de pagamento mais utilizado pelos clientes é:", metodo_pagamento_mais_utilizado)
```

```python
# Existe alguma relação entre o método de pagamento e o churn?

churn_por_metodo_pagamento = df.groupby('PaymentMethod')['Churn'].value_counts().unstack()

churn_por_metodo_pagamento['Churn_taxa'] = (churn_por_metodo_pagamento['Yes'] / (churn_por_metodo_pagamento['Yes'] + churn_por_metodo_pagamento['No'])) * 100

print("Taxa de churn por método de pagamento:")
print(churn_por_metodo_pagamento['Churn_taxa'])
```

```python
# Clientes com fatura eletrônica (PaperlessBilling) tendem a cancelar menos?

churn_por_fatura_eletronica = df.groupby('PaperlessBilling')['Churn'].value_counts().unstack()

churn_por_fatura_eletronica['Churn_taxa'] = (churn_por_fatura_eletronica['Yes'] / (churn_por_fatura_eletronica['Yes'] + churn_por_fatura_eletronica['No'])) * 100

print("Taxa de churn para clientes com e sem fatura eletrônica:")
print()
print(churn_por_fatura_eletronica['Churn_taxa'])
```
