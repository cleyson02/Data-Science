**Dados de passageiros aéreos:**

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

path = 'https://raw.githubusercontent.com/cleyson02/time-series/main/AirPassengers.csv'
dados = pd.read_csv(path)

# Converter a coluna 'Month' para o formato de data
dados['Month'] = pd.to_datetime(dados['Month'])

# Definir a coluna 'Month' como índice do DataFrame
dados.set_index('Month', inplace=True)

# Visualização dos dados
plt.figure(figsize=(10, 6))
plt.plot(dados)
plt.title('Série Temporal de Passageiros de Avião')
plt.xlabel('Ano')
plt.ylabel('Número de Passageiros')
plt.grid(True)
plt.show()

# Decomposição da série temporal
decomposicao = seasonal_decompose(dados, model='multiplicative')
tendencia = decomposicao.trend
sazonalidade = decomposicao.seasonal
residuo = decomposicao.resid

plt.figure(figsize=(10, 8))

plt.subplot(411)
plt.plot(dados, label='Série Original')
plt.legend(loc='upper left')

plt.subplot(412)
plt.plot(tendencia, label='Tendência')
plt.legend(loc='upper left')

plt.subplot(413)
plt.plot(sazonalidade, label='Sazonalidade')
plt.legend(loc='upper left')

plt.subplot(414)
plt.plot(residuo, label='Resíduo')
plt.legend(loc='upper left')

plt.tight_layout()
plt.show()

# Calculando a média e a variância ao longo do tempo
media_movel = dados.rolling(window=12).mean()
variancia_movel = dados.rolling(window=12).var()

# Plotando a média e a variância
plt.figure(figsize=(10, 6))
plt.plot(media_movel, color='red', label='Média Móvel')
plt.plot(variancia_movel, color='blue', label='Variância Móvel')
plt.title('Média e Variância Móvel ao Longo do Tempo')
plt.xlabel('Ano')
plt.ylabel('Número de Passageiros')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

path = 'https://raw.githubusercontent.com/cleyson02/Time-Series/main/BeerWineLiquor.csv'
dados = pd.read_csv(path)

# Converter a coluna 'date' para o formato de data
dados['date'] = pd.to_datetime(dados['date'])

# Definir a coluna 'date' como índice do DataFrame
dados.set_index('date', inplace=True)

# Visualização dos dados
plt.figure(figsize=(10, 6))
plt.plot(dados['beer'])
plt.title('Série Temporal de Consumo de Cerveja')
plt.xlabel('Data')
plt.ylabel('Consumo de Cerveja')
plt.grid(True)
plt.show()

# Decomposição da série temporal
decomposicao = seasonal_decompose(dados['beer'], model='multiplicative', period=12)
tendencia = decomposicao.trend
sazonalidade = decomposicao.seasonal
residuo = decomposicao.resid

plt.figure(figsize=(10, 8))

plt.subplot(411)
plt.plot(dados['beer'], label='Série Original')
plt.legend(loc='upper left')

plt.subplot(412)
plt.plot(tendencia, label='Tendência')
plt.legend(loc='upper left')

plt.subplot(413)
plt.plot(sazonalidade, label='Sazonalidade')
plt.legend(loc='upper left')

plt.subplot(414)
plt.plot(residuo, label='Resíduo')
plt.legend(loc='upper left')

plt.tight_layout()
plt.show()

# Calculando a média móvel e a variância móvel ao longo do tempo
media_movel = dados['beer'].rolling(window=30).mean()
variancia_movel = dados['beer'].rolling(window=30).var()

# Plotando a média móvel e a variância móvel
plt.figure(figsize=(10, 6))
plt.plot(media_movel, color='red', label='Média Móvel')
plt.plot(variancia_movel, color='blue', label='Variância Móvel')
plt.title('Média e Variância Móvel ao Longo do Tempo')
plt.xlabel('Data')
plt.ylabel('Valor')
plt.legend()
plt.grid(True)
plt.show()
```
